```python
import requests

url = "https://www.accuweather.com/ru/kz/almaty/222191/weather-forecast/222191"

headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

resp = requests.get(url, headers = headers)

print(resp.content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp.content,'html.parser')
print(type(soup.body.div.div.div.div))
print(soup.body)
```

```python
print(soup.div.div)
```

```python
print(soup.aside)
```

```python
for child in soup.aside.ul.children:
    print(child)
```

```python
for child in soup.body.div.descendants:
    print(child)
```

```python
print(soup.body.div.header.contents)
```
