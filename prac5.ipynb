```python
import requests
url='https://query1.finance.yahoo.com/v7/finance/download/TSLA?period1=1633396863&period2=1664932863&interval=1d&events=history&includeAdjustedClose=true'
headers={'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
resp=requests.get(url, headers = headers)
```

```python
import pandas as pd

import io

df = pd.read_csv(io.StringIO(resp.content.decode('utf-8')))
```

```python
print(df)
```

```python
l = ['TSLA', 'AAPL', 'BNGO']
for x in l:
  url='https://query1.finance.yahoo.com/v7/finance/download/'+x+'?period1=1633396863&period2=1664932863&interval=1d&events=history&includeAdjustedClose=true'
  headers={'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
  resp=requests.get(url, headers = headers)
  df = pd.read_csv(io.StringIO(resp.content.decode('utf-8')))
  print(df)
```
